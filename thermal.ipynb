```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from elements.fullModel import LumpedModel
```

```python
thermal_model = LumpedModel(T=25, dt=1)
```

```python
print(thermal_model.Tc,thermal_model.Ts)
```

```python
temps = []
temps.append([thermal_model.Tc, thermal_model.Ts])

for i in range(100):
    thermal_model.updateTemp(10,20)
    temps.append([thermal_model.Tc, thermal_model.Ts])
    print(thermal_model.Tc,thermal_model.Ts)

temps = np.array(temps)
```

```python
plt.subplot(2,1,1)
plt.plot(temps[:,0])
plt.ylabel('Core Temp')
plt.grid()
plt.subplot(2,1,2)
plt.plot(temps[:,1])
plt.ylabel('sSurface Temp')
plt.grid()
```

# Full Model

```python
from elements.fullModel import LithiumIonBattery

cell = LithiumIonBattery()
```

```python
temps = []

for i in range(100):
    cell.updateModel(100, 20)
    temps.append([cell.ECM.z_k, cell.ThM.Tc])

temps = np.array(temps).reshape(-1,2)

plt.subplot(2,1,1)
plt.plot(temps[:,0])
plt.ylabel('State of Charge')
plt.grid()
plt.subplot(2,1,2)
plt.plot(temps[:,1])
plt.ylabel('Core Temp')
plt.grid()
```
